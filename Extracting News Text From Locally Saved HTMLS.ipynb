### News Extraction from locally stored htmls

#### Extracting required tags from locally stored htmls

```python
import requests
import urllib
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
import glob
import re
```

```python
path = '/home/himalayan/htmls/'

files = [f for f in glob.glob(path + "/*.html", recursive=True)]
#print(files)
source_codes=[]
for f in files:
    html_file = open(f, 'r')
    source_code = html_file.read() 
    source_codes.append(source_code)
#print(source_codes)
titles=[]
links=[]
news=[]
for sc in source_codes:
    soup=BeautifulSoup(sc,'html.parser')
    scrapped_title=soup.title.get_text()+"\n"
    titles.append(scrapped_title)
#print(titles)
    news_each=[f for f in soup.find_all('p')]
    news_each = str(news_each)
    news_each = re.sub(r'[<p>,\n,\xa0,/,\u200d]', '',news_each)
    news_each=re.sub(r'[A-Z]', '', news_each)
    news.append(news_each)
```
